```python
from numpy import*
from sklearn import svm
from sklearn.datasets.samples_generator import make_blobs
from pylab import*
import matplotlib.pyplot 
```

```python
X, y = make_blobs(n_samples=150, centers=2, random_state=6)
scatter(X[:,0],X[:,1],c=y)
```

```python
svmm = svm.SVC(kernel='linear', C=1000)
svmm.fit(X, y)
scatter(X[:,0],X[:,1],c=y)
ax = gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
x0 = linspace(xlim[0], xlim[1], 10)
y0 = linspace(ylim[0], ylim[1], 10)
x1,y1=meshgrid(x0,y0)
xy = vstack([x1.ravel(), y1.ravel()]).T
z =svmm.decision_function(xy).reshape(y1.shape)
ax.contour(x1, y1, z, colors='k', levels=[-1,0,1], alpha=0.5,
           linestyles=['--','-','--'])
show()
```

```python
mod0 = svm.LinearSVC(C=1000)
modelo0=mod0.fit(X,y)
scatter(X[:,0],X[:,1],c=y)
ax = gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
x0 = linspace(xlim[0], xlim[1], 10)
y0 = linspace(ylim[0], ylim[1], 10)
x1,y1=meshgrid(x0,y0)
xy = vstack([x1.ravel(), y1.ravel()]).T
z =mod0.decision_function(xy).reshape(y1.shape)
ax.contour(x1, y1, z, colors='k', levels=[-1,0,1], alpha=0.5,
           linestyles=['--','-','--'])
show()
```

```python
w=modelo0.coef_[0]
print(w)
wf=-w[0]/w[1]
print(w)
xmod0=linspace(0,11,100)
ymod0=wf*xmod0-(modelo0.intercept_[0])/w[1]
scatter(X[:, 0], X[:, 1], c=y, cmap=cm.Paired)
plot(xmod0,ymod0)
```
